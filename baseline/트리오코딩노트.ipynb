```python
from io import BytesIO 
data_filename = [] 
data=[] 
i=0

with ZipFile('/content/drive/MyDrive/2022-1 인공지능 텀프로젝트/data.zip', 'r') as zipObj: 
  listOfFileNames = zipObj.namelist() 
  for fileName in listOfFileNames: 
    if fileName.endswith('csv'): 
      print(fileName) 
      zipRead = zipObj.read(fileName) 
      curr_df = pd.read_csv(BytesIO(zipRead)) 
      data.append(curr_df) 
      data_filename.append(fileName)
    i += i
```

```python
for filename, d in zip(data_filename, data): 
  print("--------------------------") 
  print(filename) 
  display(d.head(5))
```
